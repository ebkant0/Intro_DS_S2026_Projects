```python
df = pd.read_csv("adult_income_cleaned.csv")

education_income_table = df.groupby(["education", "income"]).size().unstack(fill_value=0)
education_income_sorted = education_income_table.sort_values(by=">50K", ascending=False)
print(education_income_sorted)
```

Grouping the data made it easier to compar numerical trends across categories. Variables such as age and education level seem to be connected to the ncome that an individual makes.

One thing that I found really interesting about this dataset is that it combines demographic, educational, and work related information to explore income differences. Through working with the data it became apparent that variables such as education, occupation, workclass, and amount of hours worked per week are important for understanding who earns more than $50k a year, I also noticed that categorical variables have many groups that could make grouping and visualization useful, though the dataset would need more cleaning.


```python
#data cleaning
df_clean = df.copy()
df_clean = df_clean.replace("?", pd.NA)
df_clean = df_clean.apply(lambda col: col.str.strip() if col.dtype == "object" else col)
df_clean = df_clean.drop_duplicates()
df_clean["age_group"] = pd.cut(df_clean["age"], bins=[0,24,44,64,100], labels=["18-24", "25-44", "45-64", "65+"])
print("Shape of cleaned dataset:", df_clean.shape)
print()

```

```python
print(df_clean.head())
df_clean.to_csv("adult_income_cleaned.csv", index=False)
```

```python
income_over_50k = df_clean[df_clean["income"] == ">50K"].groupby("age_group", observed=True)["income"].count()

plt.figure(figsize=(10,5))
plt.bar(income_over_50k.index.astype(str), income_over_50k.values)

plt.xlabel("Age Group")
plt.ylabel("Number of People Earning More than $50k")
plt.title("Number of People Earning More than $50k by Age Group")


plt.show()
```

```python
#histogram for age
plt.figure(figsize=(8,5))
plt.hist(df["age"], bins=20, edgecolor="black")
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

The histogram shows how age is distributed in the dataset. Most individuals are concentrated in the adult working-age range, while fewer observations appear at very low or very high ages. This suggests the dataset is centered around middle-aged adults rather than being evenly spread across all ages.

```python
#bar chart for income
income_counts = df["income"].value_counts()

plt.figure(figsize=(6,4))
income_counts.plot(kind="bar")
plt.title("Income Distribution")
plt.xlabel("Income")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.show()
```

The bar chart shows that there are more people in the lower income category than in the higher income category. This means the dataset is imbalanced, with most observations belonging to the group earning $50K or less.

```python
grouped = df.groupby("income")["education.num"].mean()

plt.figure(figsize=(6,4))
grouped.plot(kind="bar")
plt.title("Average Education Number by Income")
plt.xlabel("Income")
plt.ylabel("Average Education Number")
plt.xticks(rotation=0)
plt.show()
```

This bar plot compares the average education number for each income group. The group earning above $50K has a higher average education number than the group earning $50K or less. This suggests that higher levels of education are associated with higher income in this dataset.

```python
#scatter plot: age vs education.num
plt.figure(figsize=(8,5))
plt.scatter(df["age"], df["education.num"], alpha=0.5)
plt.title("Age vs Education Number")
plt.xlabel("Age")
plt.ylabel("Education Number")
plt.show()
```

```python
#heatmap of correlations
import seaborn as sns

plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
