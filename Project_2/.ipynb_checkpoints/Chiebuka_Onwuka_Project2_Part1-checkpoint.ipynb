```python
#filter for people earning $50k+
high_income = df[df["income"] == ">50K"]
high_income
```

```python
#filter for people with a bachelor's degree
bachelors = df[df["education"] == "Bachelors"]
bachelors
```

```python
#filter for people with a bachelor's and make more than $50k
bachelors_high_income = df[(df["education"] == "Bachelors") & (df["income"] == ">50K")]
bachelors_high_income
```

The first subset only shows people earning more than $50k. The second one shows people with a bachelor's degree and the third shows those who have both. This possibly indicates a relationship between education and high income.

```python
#data cleaning
df_clean = df.copy()
df_clean = df_clean.replace("?", pd.NA)
df_clean = df_clean.apply(lambda col: col.str.strip() if col.dtype == "object" else col)
df_clean = df_clean.drop_duplicates()
df_clean["age_group"] = pd.cut(df_clean["age"], bins=[0,24,44,64,100], labels=["18-24", "25-44", "45-64", "65+"])
print("Shape of cleaned dataset:", df_clean.shape)
print()

```
